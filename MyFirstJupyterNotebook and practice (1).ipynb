```python
sal=[44,45,46,47,48,49,50,51,52,56,57,58]
sal.sort()
sal
```

OUTLIERS

```python
import pandas as pd
ser=pd.Series(sal)
```

```python
q1=ser.quantile(0.25)
print('Q1',q1)

q3=ser.quantile(0.75)
print('Q3',q3)

iqr=q3-q1
print('IQR',iqr)
```

```python
upside=q3+(1.5*iqr)
print(upside)
```

```python
lowerside=q1-(1.5*iqr)
print(lowerside)
```
